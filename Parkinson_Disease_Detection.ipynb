```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinson_data = pd.read_excel('/content/parkinsons.xlsx')
```

```python
parkinson_data.head()
```

```python
parkinson_data.shape
```

```python
parkinson_data.info()
```

```python
parkinson_data['status'].value_counts()
```

```python
X = parkinson_data.drop(columns=['name', 'status'], axis = 1)
Y = parkinson_data['status']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,stratify = Y random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train, X_test)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, Y_train)
print('accuracy of training data: ', training_data_accuracy)
```

```python
testing_data_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(testing_data_prediction, Y_test)
print('accuracy of testing data', testing_data_accuracy)
```
